```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
ax = sns.boxplot(y='price', data=df).set_title('Price Distribution by neighbourhood group')
sns.set_theme(style='white')
plt.xlabel('neighbourhood_group')
plt.ylabel('Price')
plt.show()
```

```python
maxthresold=df['price'].quantile(0.95)
maxthresold
```

```python
df[df['price']>maxthresold]
df.head()
```

```python
minthresold=df['price'].quantile(0.05)
minthresold
```

```python
df[df['price']<minthresold]
df.head()
```

```python
df_without_outliers=df[(df['price']<maxthresold) & (df['price']>minthresold)]
df_without_outliers.shape
```

```python
df_without_outliers.describe()
```

```python
plt.figure(figsize=(10,5))
ax = sns.boxplot(y='price', data=df_without_outliers).set_title('Price Distribution')
sns.set_theme(style='white')
plt.xlabel('neighbourhood_group')
plt.ylabel('Price')
plt.show()
```

### Hence the outliers have been removed from the dataframe
